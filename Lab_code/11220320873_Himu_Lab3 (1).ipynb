```python
import matplotlib.pyplot as plt


x = [5, 5] #this is value of x
y = [0, 10] #this is value of y
plt.plot(x, y, "red", label="x = 5")
plt.fill_betweenx(y, 5, 20, color="seagreen", alpha=0.5)
plt.ylabel("Y-axis")
plt.xlabel("X-axis")
plt.title(" x >= 5 ")
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt


x = [0, 10] #this is value of x
y = [3, 3] #this is value of y
plt.plot(x, y, "red", label="y = 3")
plt.fill_between(x, y[0], color="yellow", alpha=0.5)
plt.ylabel("Y axis")
plt.xlabel("X axis")
plt.title(" y <= 3 ")
plt.grid(True)
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.legend(loc="lower right")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 5, 400) #this is value of x
y = (20 - 4 * x) / 5
plt.plot(x, y, "red", label="4x + 5y = 20")
plt.fill_between(x, y, 0,  color="violet", alpha=0.5)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Constraints & Inequalities: 4x + 5y <= 20")
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 400) #this is value of x
y = (2.5 - 0.5 * x) / 0.25
plt.plot(x, y, "red", label="0.5x + 0.25y = 2.5")
plt.fill_between(x, y, 0, color="green", alpha=0.5, where=(y >= 0))
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title(" 0.5x + 0.25y <= 2.5")
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.legend(loc="lower left")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100) #this is value of x
y = (3 * x) / 4
plt.plot(x, y, "red", label="3x - 4y = 0")
plt.fill_between(x, y, 10, where=(x * 3 / 4 <= y), color="blue", alpha=0.5)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("3x - 4y <= 0")
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.legend(loc="upper left")
plt.grid(True)
plt.show()
```
